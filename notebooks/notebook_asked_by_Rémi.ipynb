```python
import numpy as np
from GPPY.monte_carlo_test_functions import (f_2, exact_integral_f_2, support_integrands)
from GPPY.monte_carlo_tests_setup import mc_results, plot_mc_results
```

```python
r=3/4 
x = np.linspace(-r,r, 180)
X, Y = np.meshgrid(x, x)
points = np.array([X.ravel(), Y.ravel()]).T
z_f_2 = f_2(points)
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.scatter3D(X.ravel(), Y.ravel(), z_f_2, c=z_f_2)
ax.set_title(r"$f_2$")
plt.show()
```

```python
#setup
d=2
nb_sample=100
nb_core = 7
nb_point_list=np.arange(50, 450, 25).tolist()
fct_list = [f_2]
fct_names = ["f_2"]
exact_integrals= [exact_integral_f_2(d)]
support_window = support_integrands(d) 
estimators = ["MCP", "MCCV", "MCDPP",
              #"RQMC", "MC",
              ]

#mc 
if __name__ == "__main__":
    mc_results_2d, nb_point_2d = mc_results(d=d, 
                                            nb_point_list=nb_point_list, 
                                            nb_sample=nb_sample,
                                            support_window=support_window,
                                            fct_list=fct_list,
                                            fct_names=fct_names,
                                            exact_integrals=exact_integrals,
                                            estimators=estimators, 
                                            nb_core=nb_core
                                            )
```

```python
# variance

plot_mc_results(d, 
                mc_results_2d, 
                nb_point_2d, 
                fct_list=fct_list, 
                fct_names=fct_names,
                log_scale=False, 
                type_mc=estimators,
               )
```

```python
# Square Error 
plot_mc_results(d, 
                mc_list=mc_results_2d, 
                nb_point_list=nb_point_2d, 
                fct_list=fct_list, 
                fct_names=fct_names,
                error_type="SE",
                plot_error=True,
                plot_std=False,
                log_scale=True,
                nb_subsample_nb_points=5,
                type_mc=estimators,
               )
```
